```python
from bs4 import BeautifulSoup as bs
from splinter import Browser
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
url = 'https://www.bloodhorse.com/horse-racing/racetracks'
browser.visit(url)
html = browser.html
soup = bs(html, 'html.parser')
```

```python
soup
```

```python
unordered_list = soup.find('ul', class_="fixed-column-list")
track_list = unordered_list.find_all('li')
track_list
```

```python
lat = []
lon = []
nameShort = []
nameLong = []

for t in track_list:
    t_lat = t['data-latitude']
    lat.append(t_lat)
    t_lon = t['data-longitude']
    lon.append(t_lon)
    t_nameLong = t.a.text.replace("\n                    ","")
    nameLong.append(t_nameLong)
    t_nameShort = t['id']
    nameShort.append(t_nameShort)
print(f"""
Lat
{lat}
--------------------------------------------------------------------
Lon
{lon}
--------------------------------------------------------------------
nameLong
{nameLong}
--------------------------------------------------------------------
nameShort
{nameShort}
""")
```
